Step 1: Load Dataset

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#file loading
iris = pd.read_csv('IRIS.csv')
```

```python
#viewing the first five rows
iris.head()
```

Step 2: Data Cleaning and Checking Data Properties

```python
# Check for null values and basic info
print("Dataset Information:")
iris.info()
```

```python
print("\nMissing Values Count:")
print(iris.isnull().sum())
```

```python
print("\nStatistical Summary:")
display(iris.describe())
```

Step 3: Data Encoding

```python
# Strip spaces and map species to numbers
iris['species'] = iris['species'].str.strip().map({
    'Iris-setosa': 0, 
    'Iris-versicolor': 1, 
    'Iris-virginica': 2
})
```

```python
#check if any NaNs were created during mapping
if iris['species'].isnull().any():
    iris.dropna(subset=['species'], inplace=True)
    print("Dropped rows with missing or unmapped species labels.")
```

```python
iris.head()
```

Step 4: Split Data

```python
# Define Features and Target
X = iris.drop('species', axis=1)
y = iris['species'].astype(int)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training set size: {X_train.shape[0]}")
print(f"Testing set size: {X_test.shape[0]}")
```

Step 5: Model Training and Evaluation

```python
# Define the models
models = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "Random Forest Classifier": RandomForestClassifier(random_state=42),
    "Gradient Boosting Classifier": GradientBoostingClassifier(random_state=42),
    "K-Nearest Neighbors (KNN)": KNeighborsClassifier(n_neighbors=5),
    "Support Vector Machine (SVM)": SVC(kernel='linear')
}
```

```python
# Training and Evaluation Loop
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    print(f"\n--- {name} ---")
    print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
```

## Step 6: Results and Best Models
After evaluating all five models, the results are as follows:
 **Logistic Regression**: 100% Accuracy
 **Random Forest**: 100% Accuracy
 **Gradient Boosting**: 100% Accuracy
 **K-Nearest Neighbors (KNN)**: 100% Accuracy
 **Support Vector Machine (SVM)**: 100% Accuracy

### Analysis:
All models achieved perfect accuracy with zero misclassifications in their respective confusion matrices. This indicates that the features in the Iris dataset are highly separable. For this specific task, any of these models would be an excellent choice, though **Logistic Regression** and **SVM** are particularly efficient for this type of linear classification.
